### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python

median_mpg = cars_sheet['mpg'].median()
above_median_mpg_cars = cars_sheet[cars_sheet['mpg'] > median_mpg].copy()


above_median_mpg_cars.reset_index(inplace=True)

best_ratio_cars = above_median_mpg_cars[above_median_mpg_cars['hp/weight'] == above_median_mpg_cars['hp/weight'].max()]

print("Car(s) with the best horsepower to weight ratio among cars with above-median mpg:")
print(best_ratio_cars[['horsepower', 'weight', 'mpg', 'hp/weight']])
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python

avg_mpg_by_year_origin = cars_sheet.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(data=avg_mpg_by_year_origin, x='model_year', y='mpg', hue='origin')

plt.title('Average MPG Evolution by Year and Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.grid(True)
plt.legend(title='Origin')
plt.show()
```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
car_appearances = cars_sheet.reset_index().groupby('name')['model_year'].nunique()
multi_year_cars = car_appearances[car_appearances > 1].index
multi_year_df = cars_sheet.reset_index()[cars_sheet.reset_index()['name'].isin(multi_year_cars)]
mpg_stats = multi_year_df.groupby('name')['mpg'].agg(['std', 'mean', 'count'])
consistent_cars = mpg_stats[(mpg_stats['std'] < 1.0) & (mpg_stats['count'] >= 2)]

if not consistent_cars.empty:
    lowest_std = consistent_cars['std'].min()
    lowest_variation_cars = consistent_cars[consistent_cars['std'] == lowest_std]

    # Sort the result
    sorted_result = lowest_variation_cars.sort_values(by=['count', 'mean'], ascending=[False, False])

    print("Car(s) with the lowest variation in mpg (std < 1.0 and >= 2 appearances):")
    print(sorted_result[['count', 'mean']])
else:
    print("No consistent car models found with the specified criteria.")
```
